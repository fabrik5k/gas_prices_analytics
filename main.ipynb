### 12. Você conseguiria obter uma tabela contendo a variação percentual ano a ano para o estado do Rio de Janeiro?

### DESAFIO: Crie uma tabela contendo uma serie temporal com a diferença absoluta e percentual entre os valores mais baratos e caros.

```python
tabela_st_da = pd.DataFrame()
amplitude_por_estado = {}

#Pegando todos os dados onde o PRODUTO é a gasolina comum
gas_comum = gas_concat[gas_concat['PRODUTO'] == 'GASOLINA COMUM']
gas_comum = gas_comum.sort_values(by='DATA FINAL')

#Estou fazendo isso porque eu quero pegar a variação percentual e valor absoluto de cada estado em relação a ele mesmo para que aja menos interfereência de outros dados possiveis
for estado in gas_comum['ESTADO'].unique():
    gas_estado = gas_comum[gas_comum['ESTADO'] == estado]

    var_percentual_med = gas_estado['PREÇO MÉDIO REVENDA'].pct_change() * 100
    var_absoluta_med = gas_estado['PREÇO MÉDIO REVENDA'].diff()
    amplitude_med = gas_estado['PREÇO MÉDIO REVENDA'].max() - gas_estado['PREÇO MÉDIO REVENDA'].min()

    var_percentual_min = gas_estado['PREÇO MÍNIMO REVENDA'].pct_change() * 100
    var_absoluta_min = gas_estado['PREÇO MÍNIMO REVENDA'].diff()
    amplitude_min = gas_estado['PREÇO MÍNIMO REVENDA'].max() - gas_estado['PREÇO MÍNIMO REVENDA'].min()

    var_percentual_max = gas_estado['PREÇO MÁXIMO REVENDA'].pct_change() * 100
    var_absoluta_max = gas_estado['PREÇO MÁXIMO REVENDA'].diff()
    amplitude_max = gas_estado['PREÇO MÁXIMO REVENDA'].max() - gas_estado['PREÇO MÁXIMO REVENDA'].min()

    amplitude_por_estado[str(estado)] = round(amplitude_med, 2)

    temp_df = pd.DataFrame({
        'ESTADO': estado,
        'PREÇO MÉDIO REVENDA': gas_estado['PREÇO MÉDIO REVENDA'],
        'PREÇO MÍNIMO REVENDA': gas_estado['PREÇO MÍNIMO REVENDA'],
        'PREÇO MÁXIMO REVENDA': gas_estado['PREÇO MÁXIMO REVENDA'],
        'DATA FINAL': gas_estado['DATA FINAL'],
        'VARIAÇÃO PERCENTUAL MÉDIA REVENDA': var_percentual_med,
        'DIFERENÇA ABSOLUTA MÉDIA REVENDA': var_absoluta_med,
        'VARIAÇÃO PERCENTUAL MÍNIMO REVENDA': var_percentual_min,
        'DIFERENÇA ABSOLUTA MÍNIMO REVENDA': var_absoluta_min,
        'VARIAÇÃO PERCENTUAL MÁXIMO REVENDA': var_percentual_max,
        'DIFERENÇA ABSOLUTA MÁXIMO REVENDA': var_absoluta_max
    })

    tabela_st_da = pd.concat([tabela_st_da, temp_df], ignore_index=True)

tabela_st_da
```

```python
para = tabela_st_da[tabela_st_da['ESTADO'] == 'PARA'].sort_values(by='DATA FINAL')
plt.figure(figsize=(10, 6))
plt.plot(para['DATA FINAL'], para['PREÇO MÉDIO REVENDA'])
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())

plt.gcf().autofmt_xdate()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.plot(para['DATA FINAL'], para['VARIAÇÃO PERCENTUAL MÉDIA REVENDA'])
plt.gca().xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
plt.gca().xaxis.set_major_locator(mdates.YearLocator())

plt.gcf().autofmt_xdate()
plt.show()
```
